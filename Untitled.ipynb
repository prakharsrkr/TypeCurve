```python
import numpy as np

def validate_inputs(qi, di, b, Dlim, IBU, MBU):
    # A simple validation function where IBU and MBU can be zero
    return all(x >= 0 for x in [IBU, MBU]) and all(x > 0 for x in [qi, di, b, Dlim])
    
def modified_hyperbolic(time_array, qi, di, b, Dlim, IBU, MBU, buildup_method='Linear', epsilon=1e-10):
    """
    Calculate the monthly production volume using the Modified Hyperbolic Decline Model with a buildup phase.
    """
    # Validate inputs
    if not validate_inputs(qi, di, b, Dlim, IBU, MBU):
        #print(f"Invalid inputs detected: qi={qi}, di={di}, b={b}, Dlim={Dlim}, IBU={IBU}, MBU={MBU}")
        return np.zeros_like(time_array)
    
    # Convert decline rates to nominal rates
    try:
        di_nominal = ((1 - 0.01 * di) ** (-b) - 1) / (12 * b)
        Dlim_nominal = ((1 - 0.01 * Dlim) ** (-b) - 1) / (12 * b)
    except (ZeroDivisionError, OverflowError, ValueError) as e:
        #print(f"Invalid decline rates: di={di}, Dlim={Dlim}, b={b}. Error: {e}")
        return np.zeros_like(time_array)

    # Calculate switch point
    try:
        qlim = qi * (Dlim_nominal / (di_nominal + epsilon)) ** (1.0 / b)
        tlim = ((qi / qlim) ** b - 1.0) / (b * (di_nominal + epsilon))
    except (ZeroDivisionError, OverflowError, ValueError) as e:
        # print(f"Invalid switch point calculations: qi={qi}, qlim={qlim}, di={di}, Dlim={Dlim}, b={b}. Error: {e}")
        return np.zeros_like(time_array)

    # Handle the case when MBU (buildup period) is zero
    if MBU == 0:
        # If MBU is zero, skip the buildup phase and start with the hyperbolic decline directly
        t_buildup = np.array([])
        buildup_production = np.array([])
        t_post_buildup = time_array
        t_hyp = t_post_buildup[t_post_buildup < tlim]
        t_exp = t_post_buildup[t_post_buildup >= tlim]
    else:
        # Separate time into buildup, hyperbolic, and exponential segments
        t_buildup = time_array[time_array <= MBU]
        t_post_buildup = time_array[time_array > MBU]
        t_hyp = t_post_buildup[t_post_buildup < tlim]
        t_exp = t_post_buildup[t_post_buildup >= tlim]

        # Calculate production during buildup period
        if buildup_method == 'Flat':
            buildup_production = IBU + np.zeros_like(t_buildup)
        elif buildup_method == 'Linear':
            slope = (qi - IBU) / (MBU + epsilon)
            buildup_production = IBU + slope * t_buildup
        elif buildup_method == 'Exp':
            slope = np.log(qi / (IBU + epsilon)) / (MBU + epsilon)
            buildup_production = IBU * np.exp(slope * t_buildup)
        else:
            raise ValueError(f"Unknown buildup method: {buildup_method}")

    # Calculate production during hyperbolic decline period
    try:
        q_model_hyp = qi * (1.0 + b * di_nominal * (t_hyp - MBU)) ** (-1.0 / b)
    except (ZeroDivisionError, OverflowError, ValueError) as e:
        # print(f"Invalid hyperbolic decline calculation: qi={qi}, di={di}, b={b}. Error: {e}")
        q_model_hyp = np.zeros_like(t_hyp)

    # Calculate production during exponential decline period
    try:
        q_model_exp = qlim * np.exp(-Dlim_nominal * (t_exp - tlim))
    except (ZeroDivisionError, OverflowError, ValueError) as e:
        # print(f"Invalid exponential decline calculation: qlim={qlim}, Dlim={Dlim}, tlim={tlim}. Error: {e}")
        q_model_exp = np.zeros_like(t_exp)

    # Smooth transition at tlim (Hyperbolic to Exponential)
    if len(t_exp) > 0 and len(t_hyp) > 0:
        time_fraction = (tlim - t_hyp[-1]) / (t_exp[0] - t_hyp[-1] + epsilon)
        q_model_exp[0] = time_fraction * q_model_hyp[-1] + (1 - time_fraction) * q_model_exp[0]

    # Combine production rates
    production_post_buildup = np.concatenate((q_model_hyp, q_model_exp))
    daily_production = np.concatenate((buildup_production, production_post_buildup))

    # Calculate monthly production volumes by summing daily rates over each month
    days_per_month = 30
    monthly_volumes = daily_production * days_per_month

    return daily_production, monthly_volumes

# Example usage
time_array = np.linspace(0, 120, 121)  # 10 years, monthly steps
qi = 16141.6435546875  # Initial rate
di = 106.9966812133789 # Initial decline rate (percent)
b = 0.8006053566932678  # b-factor
Dlim = 7  # Limiting decline rate (percent)
IBU = 0  # Initial buildup rate
MBU = 0  # Buildup period (months)

daily_production, monthly_volumes = modified_hyperbolic(time_array, qi, di, b, Dlim, IBU, MBU)
import matplotlib.pyplot as plt

def plot_production(time_array, daily_production, monthly_volumes):
    """
    Plot daily production and monthly volumes.

    Parameters:
    - time_array: Array of time values (in months).
    - daily_production: Array of daily production rates (in bbls/day).
    - monthly_volumes: Array of monthly production volumes (in bbls/month).
    """
    # Convert time to years for better readability on the x-axis
    time_years = time_array / 12

    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot daily production
    ax1.plot(time_years, daily_production, 'b-', label='Daily Production (bbls/day)')
    ax1.set_xlabel('Time (years)')
    ax1.set_ylabel('Daily Production (bbls/day)', color='b')
    ax1.tick_params(axis='y', labelcolor='b')

    # Create a secondary y-axis for the monthly volumes
    ax2 = ax1.twinx()
    months = np.arange(len(monthly_volumes))
    ax2.bar(months / 12, monthly_volumes, color='orange', width=0.1, alpha=0.6, label='Monthly Volumes (bbls/month)')
    ax2.set_ylabel('Monthly Volumes (bbls/month)', color='orange')
    ax2.tick_params(axis='y', labelcolor='orange')

    # Add grid, title, and legend
    ax1.grid(True)
    fig.suptitle('Daily Production and Monthly Volumes Over Time')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')

    plt.show()

def cumulative_production_and_EUR(qi, di, b, Dlim, IBU, MBU, years=30, buildup_method='Linear'):
    months = np.linspace(0, years * 12, years * 12 + 1)
    production = modified_hyperbolic(months, qi, di, b, Dlim, IBU, MBU, buildup_method)[0]
    cumulative_production = np.cumsum(production)
    EUR = cumulative_production[-1]
    print(EUR)
    return cumulative_production, EUR

def plot_production(time_array, daily_production, monthly_volumes):
    """
    Plot daily production and monthly volumes in two separate plots.

    Parameters:
    - time_array: Array of time values (in months).
    - daily_production: Array of daily production rates (in bbls/day).
    - monthly_volumes: Array of monthly production volumes (in bbls/month).
    """
    # Convert time to years for better readability on the x-axis
    time_years = time_array / 12
    months_years = np.arange(len(monthly_volumes)) / 12

    # Plot for Daily Production
    plt.figure(figsize=(10, 6))
    plt.plot(time_years, daily_production, 'b-', label='Daily Production (bbls/day)')
    plt.xlabel('Time (years)')
    plt.ylabel('Daily Production (bbls/day)')
    plt.title('Daily Production Over Time')
    plt.yscale('log')
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()

    # Plot for Monthly Volumes
    plt.figure(figsize=(10, 6))
    plt.bar(months_years, monthly_volumes, color='orange', width=0.1, alpha=0.6, label='Monthly Volumes (bbls/month)')
    plt.xlabel('Time (years)')
    plt.ylabel('Monthly Volumes (bbls/month)')
    plt.title('Monthly Volumes Over Time')
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()

# Example usage
time_array = np.linspace(0, 12*30, 12*30+1)  # 10 years, monthly steps
qi = 33708  # Initial rate
di = 77.9  # Initial decline rate (percent)
b = 1.2  # b-factor
Dlim = 7  # Limiting decline rate (percent)
IBU = 1000  # Initial buildup rate
MBU = 2  # Buildup period (months)

daily_production, monthly_volumes = modified_hyperbolic(time_array, qi, di, b, Dlim, IBU, MBU)
plot_production(time_array, daily_production, monthly_volumes)
```

```python
import os
import re

BASE_DIR = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\URTeC2024_Papers"

def format_folder_name(folder_name):
    """Format the folder name to include 'Theme' before the number and a hyphen after the number, removing excessive spaces."""
    # Remove excessive spaces
    folder_name = re.sub(r'\s+', ' ', folder_name).strip()
    
    # Check if 'Theme' is already included and match the pattern
    match = re.match(r'(Theme\s+)?(\d+)\s*-?\s*(.*)', folder_name, re.IGNORECASE)
    if match:
        theme_prefix, number, theme = match.groups()
        formatted_name = f"Theme {number} - {theme.strip()}"
        # Remove trailing dash if present
        formatted_name = formatted_name.rstrip(' -') if formatted_name.endswith(' -') else formatted_name
        print(f"Matched existing 'Theme' pattern: {folder_name} -> {formatted_name}")
        return formatted_name
    else:
        # If no match, assume the folder name is just a number or number with spaces
        match = re.match(r'(\d+)', folder_name)
        if match:
            number = match.group(1)
            formatted_name = f"Theme {number} -"
            print(f"Matched number pattern: {folder_name} -> {formatted_name}")
            return formatted_name
        else:
            print(f"No match found for: {folder_name}")
            return folder_name  # Return as is if it doesn't match the expected pattern

def rename_folders(base_dir):
    for folder_name in os.listdir(base_dir):
        old_path = os.path.join(base_dir, folder_name)
        if os.path.isdir(old_path):
            new_folder_name = format_folder_name(folder_name)
            new_path = os.path.join(base_dir, new_folder_name)
            if old_path != new_path:  # Ensure we don't try to rename to the same name
                try:
                    os.rename(old_path, new_path)
                    print(f"Renamed {old_path} to {new_path}")
                except Exception as e:
                    print(f"Failed to rename {old_path} to {new_path}: {e}")
            else:
                print(f"No rename needed for {old_path}")

def main():
    rename_folders(BASE_DIR)
    print("Folder renaming completed.")

if __name__ == "__main__":
    main()
```

```python
HTML_FILE = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\new 8.html"

# Base directory path
BASE_DIR = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\URTeC2024_Papers"


def parse_html(html_file):
    """Parse the HTML file to create a mapping of file IDs to titles."""
    with open(html_file, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
    
    id_to_title = {}
    for link in soup.find_all('a', href=True):
        if 'extended-abstracts' in link['href']:
            file_id = link['href'].split('/')[-1].split('.')[0]
            title = link.get_text(strip=True)
            id_to_title[file_id] = title
    return id_to_title

def sanitize_filename(filename, max_length=50):
    """Replace special characters and spaces in filenames and truncate if necessary."""
    filename = re.sub(r'[<>:"/\\|?*]', '_', filename)
    if len(filename) > max_length:
        filename = filename[:max_length]
    return filename

def shorten_folder_names(base_dir):
    """Shorten folder names to reduce the overall path length."""
    for root, dirs, files in os.walk(base_dir):
        for dir_name in dirs:
            short_name = sanitize_filename(dir_name, max_length=30)
            old_path = os.path.join(root, dir_name)
            new_path = os.path.join(root, short_name)
            if old_path != new_path:
                os.rename(old_path, new_path)

def rename_files_in_folders(base_dir, id_to_title):
    """Rename files in folders based on the ID to title mapping."""
    id_pattern = re.compile(r'^\d+$')  # Pattern to check if filename is still an ID
    for root, dirs, files in os.walk(base_dir):
        for filename in files:
            file_id = filename.split('.')[0]
            if id_pattern.match(file_id) and file_id in id_to_title:
                sanitized_title = sanitize_filename(id_to_title[file_id], max_length=100)
                new_filename = f"{sanitized_title}.pdf"
                old_path = os.path.join(root, filename)
                new_path = os.path.join(root, new_filename)
                
                if os.path.exists(old_path):
                    try:
                        if len(new_path) > 260:
                            print(f"Path length exceeds limit: {new_path}")
                        else:
                            os.rename(old_path, new_path)
                            print(f"Renamed {old_path} to {new_path}")
                    except Exception as e:
                        print(f"Failed to rename {old_path} to {new_path}: {e}")
                else:
                    print(f"File does not exist: {old_path}")

def main():
    id_to_title = parse_html(HTML_FILE)
    shorten_folder_names(BASE_DIR)
    rename_files_in_folders(BASE_DIR, id_to_title)
    print("File renaming completed.")

if __name__ == "__main__":
    main()
```
