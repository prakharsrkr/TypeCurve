```python
import os
import shutil
from bs4 import BeautifulSoup
from difflib import get_close_matches

# Paths
HTML_FILE_PATH = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\new 8.html"
BASE_DIR = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\URTeC2024_Papers"

def sanitize_filename(filename):
    """Sanitize the filename to be valid on all operating systems."""
    return filename.replace(':', ' -').replace('/', '_').replace('\\', '_').replace('?', '').replace('*', '').replace('"', '').replace('<', '').replace('>', '').replace('|', '').replace(',', '').replace('-', ' ').strip()

def extract_details_from_html(html_file_path):
    with open(html_file_path, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')

    details = []

    for row in soup.find_all('tr', id=lambda x: x and x.startswith('row')):
        try:
            control_id = row['id'][3:]
            link = row.find('a', id=lambda x: x and x.startswith('link'))
            title = link.find('b').text.strip()
            session = row.find('td', class_='description')['data-session']

            # Extract the theme name properly
            theme = ' '.join(session.split(' ')[1:]).strip()
            details.append({
                'id': control_id,
                'theme': theme,
                'title': title
            })
        except Exception as e:
            print(f"Error processing row: {e}")

    return details

def get_closest_theme_dir(base_dir, theme):
    """Get the closest matching directory name to the theme."""
    existing_dirs = [d for d in os.listdir(base_dir) if os.path.isdir(os.path.join(base_dir, d))]
    if not existing_dirs:
        return None
    closest_match = get_close_matches(sanitize_filename(theme), existing_dirs, n=1)
    return closest_match[0] if closest_match else None

def move_and_rename_files(details, base_dir):
    failed_moves = []

    for detail in details:
        original_file = os.path.join(base_dir, f"{detail['id']}.pdf")
        if not os.path.exists(original_file):
            print(f"File {original_file} does not exist")
            continue

        theme_dir = os.path.join(base_dir, sanitize_filename(detail['theme']))
        if not os.path.exists(theme_dir):
            closest_theme_dir = get_closest_theme_dir(base_dir, detail['theme'])
            if closest_theme_dir:
                theme_dir = os.path.join(base_dir, closest_theme_dir)
            else:
                try:
                    os.makedirs(theme_dir)
                except OSError as e:
                    print(f"Error creating directory {theme_dir}: {e}")
                    failed_moves.append((original_file, detail['theme'], "Directory creation failed"))
                    continue

        new_filename = sanitize_filename(detail['title'][:100]) + ".pdf"  # Truncate to keep the filename short
        new_filepath = os.path.join(theme_dir, new_filename)

        try:
            shutil.move(original_file, new_filepath)
            print(f"Moved {original_file} to {new_filepath}")
        except Exception as e:
            print(f"Failed to move {original_file} to {new_filepath}: {e}")
            failed_moves.append((original_file, detail['theme'], str(e)))

    if failed_moves:
        print("\nFailed Moves:")
        for original_file, theme, reason in failed_moves:
            print(f"Failed to move {original_file} to theme '{theme}': {reason}")

    return failed_moves

def handle_failed_moves(failed_moves):
    for original_file, theme, reason in failed_moves:
        print(f"\nFailed to move {original_file} to theme '{theme}': {reason}")
        existing_dirs = [d for d in os.listdir(BASE_DIR) if os.path.isdir(os.path.join(BASE_DIR, d))]
        closest_match = get_close_matches(sanitize_filename(theme), existing_dirs, n=1)
        if closest_match:
            print(f"Suggested closest match: {closest_match[0]}")
            user_input = input(f"Do you want to move the file to {closest_match[0]}? (y/n): ")
            if user_input.lower() == 'y':
                new_theme_dir = os.path.join(BASE_DIR, closest_match[0])
                new_filename = os.path.basename(original_file)
                new_filepath = os.path.join(new_theme_dir, new_filename)
                try:
                    shutil.move(original_file, new_filepath)
                    print(f"Moved {original_file} to {new_filepath}")
                except Exception as e:
                    print(f"Failed to move {original_file} to {new_filepath}: {e}")
            else:
                print(f"Skipped moving {original_file}")
        else:
            print("No close match found. Skipping.")

def main():
    details = extract_details_from_html(HTML_FILE_PATH)
    failed_moves = move_and_rename_files(details, BASE_DIR)
    handle_failed_moves(failed_moves)
    print("Completed processing all files.")

if __name__ == "__main__":
    main()
```

```python
import os
import re

BASE_DIR = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\URTeC2024_Papers"

def format_folder_name(folder_name):
    """Format the folder name to include 'Theme' before the number and a hyphen after the number, removing excessive spaces."""
    # Remove excessive spaces
    folder_name = re.sub(r'\s+', ' ', folder_name).strip()
    
    # Check if 'Theme' is already included and match the pattern
    match = re.match(r'(Theme\s+)?(\d+)\s*-?\s*(.*)', folder_name, re.IGNORECASE)
    if match:
        theme_prefix, number, theme = match.groups()
        formatted_name = f"Theme {number} - {theme.strip()}"
        # Remove trailing dash if present
        formatted_name = formatted_name.rstrip(' -') if formatted_name.endswith(' -') else formatted_name
        print(f"Matched existing 'Theme' pattern: {folder_name} -> {formatted_name}")
        return formatted_name
    else:
        # If no match, assume the folder name is just a number or number with spaces
        match = re.match(r'(\d+)', folder_name)
        if match:
            number = match.group(1)
            formatted_name = f"Theme {number} -"
            print(f"Matched number pattern: {folder_name} -> {formatted_name}")
            return formatted_name
        else:
            print(f"No match found for: {folder_name}")
            return folder_name  # Return as is if it doesn't match the expected pattern

def rename_folders(base_dir):
    for folder_name in os.listdir(base_dir):
        old_path = os.path.join(base_dir, folder_name)
        if os.path.isdir(old_path):
            new_folder_name = format_folder_name(folder_name)
            new_path = os.path.join(base_dir, new_folder_name)
            if old_path != new_path:  # Ensure we don't try to rename to the same name
                try:
                    os.rename(old_path, new_path)
                    print(f"Renamed {old_path} to {new_path}")
                except Exception as e:
                    print(f"Failed to rename {old_path} to {new_path}: {e}")
            else:
                print(f"No rename needed for {old_path}")

def main():
    rename_folders(BASE_DIR)
    print("Folder renaming completed.")

if __name__ == "__main__":
    main()
```

```python
HTML_FILE = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\new 8.html"

# Base directory path
BASE_DIR = r"C:\Users\Prakhar.Sarkar\OneDrive - SRP Management Services\Documents\_For_Prakhar\URTeC2024_Papers"


def parse_html(html_file):
    """Parse the HTML file to create a mapping of file IDs to titles."""
    with open(html_file, 'r', encoding='utf-8') as file:
        soup = BeautifulSoup(file, 'html.parser')
    
    id_to_title = {}
    for link in soup.find_all('a', href=True):
        if 'extended-abstracts' in link['href']:
            file_id = link['href'].split('/')[-1].split('.')[0]
            title = link.get_text(strip=True)
            id_to_title[file_id] = title
    return id_to_title

def sanitize_filename(filename, max_length=50):
    """Replace special characters and spaces in filenames and truncate if necessary."""
    filename = re.sub(r'[<>:"/\\|?*]', '_', filename)
    if len(filename) > max_length:
        filename = filename[:max_length]
    return filename

def shorten_folder_names(base_dir):
    """Shorten folder names to reduce the overall path length."""
    for root, dirs, files in os.walk(base_dir):
        for dir_name in dirs:
            short_name = sanitize_filename(dir_name, max_length=30)
            old_path = os.path.join(root, dir_name)
            new_path = os.path.join(root, short_name)
            if old_path != new_path:
                os.rename(old_path, new_path)

def rename_files_in_folders(base_dir, id_to_title):
    """Rename files in folders based on the ID to title mapping."""
    id_pattern = re.compile(r'^\d+$')  # Pattern to check if filename is still an ID
    for root, dirs, files in os.walk(base_dir):
        for filename in files:
            file_id = filename.split('.')[0]
            if id_pattern.match(file_id) and file_id in id_to_title:
                sanitized_title = sanitize_filename(id_to_title[file_id], max_length=100)
                new_filename = f"{sanitized_title}.pdf"
                old_path = os.path.join(root, filename)
                new_path = os.path.join(root, new_filename)
                
                if os.path.exists(old_path):
                    try:
                        if len(new_path) > 260:
                            print(f"Path length exceeds limit: {new_path}")
                        else:
                            os.rename(old_path, new_path)
                            print(f"Renamed {old_path} to {new_path}")
                    except Exception as e:
                        print(f"Failed to rename {old_path} to {new_path}: {e}")
                else:
                    print(f"File does not exist: {old_path}")

def main():
    id_to_title = parse_html(HTML_FILE)
    shorten_folder_names(BASE_DIR)
    rename_files_in_folders(BASE_DIR, id_to_title)
    print("File renaming completed.")

if __name__ == "__main__":
    main()
```
